```python
import networkx as nx
```

```python
from P3.utils import read_network, draw_network
```

## Exploring layouts

```python
G = read_network('..\\P2\\soc-dolphins.mtx')
```

```python
draw_network(G, "Dolphins")
```

```python
pos = nx.kamada_kawai_layout(G)
draw_network(G, "Dolphins", pos=pos)
```

```python
pos = nx.spectral_layout(G)
draw_network(G, "Dolphins", pos=pos)
```

```python
pos = nx.spiral_layout(G)
draw_network(G, "Dolphins", pos=pos)
```

```python
pos = nx.arf_layout(G)
draw_network(G, "Dolphins", pos=pos)
```

```python
pos = nx.fruchterman_reingold_layout(G)
draw_network(G, "Dolphins", pos=pos)
```
