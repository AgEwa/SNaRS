```python
from P10.spectral import detect_communities_from_csv, visualize_partition_clustered, save_node2comm_to_csv
```

```python
files = ["D1-K=2", "D2-K=7", "D3-K=12", "D3-UNC"]
for file in files:
    n_comm = int(file.split("K=")[1]) if "K=" in file else None

    A, labels, k, elapsed = detect_communities_from_csv(f"competition/{file}.csv", has_header=False, k=n_comm)

    visualize_partition_clustered(A, labels, title=f"{file}: Spectral Clusters (k={k})")

    with open("description.txt", "a", encoding="utf-8") as f:
        f.write(f"{file}.csv,{elapsed:.4f}\n")

    save_node2comm_to_csv(labels, f"{file}.csv")
```

based on algorithm from paper (with adjacency matrix as affinity matrix):
Ng, A. et al. On Spectral Clustering: Analysis and an algorithm. Neural Information Processing Systems (2001).
